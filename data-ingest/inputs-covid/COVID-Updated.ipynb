```python
import pandas as pd
import numpy as np
import datetime
```

```python
covid = pd.read_csv ('/Users/kelseyrobb/Desktop/COVID/Johns Hopkins Data/time_series_19-covid-Confirmed-3-8.csv', index_col = False)
```

```python
covid.head()
```

```python
dates = covid.loc[:, covid.columns.str.startswith(('1', '2', '3'))]
```

```python
dates.head()
```

```python
dates = dates.diff(axis=1)
```

```python
dates.head()
```

```python
dates = dates.drop(columns = ['1/22/20'])
```

```python
dates.head()
```

```python
covid_new = covid[['Province/State', 'Country/Region', 'Lat', 'Long', '1/22/20']]
```

```python
covid_new.head()
```

```python
combine = pd.concat([covid_new, dates], axis=1)
```

```python
combine.head()
```

```python
combine['Country/Region'].replace(['Mainland China', 'US', 'Macau', 'Hong Kong', 'UK', 'North Macedonia', 'Vatican City', 'Republic of Ireland', 'Serbia'],['China', 'United States of America', 'Macao S.A.R.', 'Hong Kong S.A.R.', 'United Kingdom', 'Macedonia', 'Vatican', 'Ireland', 'Republic of Serbia'], inplace=True)
```

```python
combine.head()
```

```python
combine.drop(combine[(combine['Province/State'] == 'Travis, CA (From Diamond Princess)') | (combine['Province/State'] == 'Lackland, TX (From Diamond Princess)') | (combine['Province/State'] == 'Omaha, NE (From Diamond Princess)')].index, inplace=True)
```

```python
combine.head(-20)
```

```python
country_sum = combine.groupby('Country/Region', as_index=False).sum()
country_sum.head()
```

```python
country_sum['Country/Region'].replace(['Others'],['Diamond Princess Cruise Ship'], inplace=True)
```

```python
combine_new = combine[(combine['Country/Region'] == 'United States of America') | (combine['Country/Region'] == 'Canada') | (combine['Country/Region'] == 'Australia') | (combine['Country/Region'] == 'China') | (combine['Country/Region'] == 'France')]
```

```python
combine_new.head(-20)
```

```python
combine_new = combine_new.drop(['Country/Region'], axis=1)
```

```python
combine_new.rename({'Province/State': 'Country/Region'}, axis=1, inplace=True)
```

```python
combine_new.head()
```

```python
combine_new.drop(combine_new[(combine_new['Country/Region'] == 'From Diamond Princess') | (combine_new['Country/Region'] == 'Unassigned Location (From Diamond Princess)')].index, inplace=True)
```

```python
combine_new.head(-10)
```

```python
final = combine_new.append(country_sum)
```

```python
print(final)
```

```python
final['Country/Region'] = final['Country/Region'].str.split(',').str[0]
```

```python
final.head(-40)
```

```python
final.dropna(subset = ["Country/Region"], inplace=True)
```

```python
final.set_index('Country/Region', inplace=True)
```

```python
final.to_csv('/Users/kelseyrobb/Documents/GitHub/measles/data-ingest/inputs-covid/daily_covid_counts-3-9.csv')
```
